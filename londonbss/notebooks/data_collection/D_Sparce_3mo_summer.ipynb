```python
def read_url(url):
    s=requests.get(url).content
    c=pd.read_csv(io.StringIO(s.decode('utf-8')))
    return c
```

```python
frames = []
for x in data['usage-stats/']:
    frames.append(read_url(data_path + x))

result = pd.concat(frames)
result.columns
```

```python
import tensorly as tl
from tensorly.decomposition import tucker
```

```python
#
```

```python
# Rank of Tucker decomposition
tucker_rank = [25, 25, 30]
```

```python
core, factors = tucker(tensor, rank=tucker_rank, init='random', tol=10e-5,random_state=12345)
```

```python
import tensorflow as tf
```

```python
type(core)
```

```python
core_tensor=tf.convert_to_tensor(core)
dataset_core_tensor=tf.data.Dataset.from_tensors(core_tensor)
path_core='london-bss-colab/core_folder_3mo_summer'
dataset_core_tensor.save(path_core)
```

```python
path_factors='london-bss-colab/factors_folder_3mo_summer/'
# os.makedirs(path_factors)
for i,factor in enumerate(factors):
  df=pd.DataFrame(factor)
  df.to_csv(path_factors+str(i)+'.csv')
```
