## 3. Encode location zone

```python
all_events_df.set_index("title", inplace=True)
```

```python
all_events_df["Latitude"].value_counts()
```

```python
def make_floats(x):
    print(x)
    if x == "London-wide":
        x = "London-wide"
    elif type(x) == str:
        x = float(x)
    else:
        x = 0
    return x
```

```python
all_events_df['Latitude'] = all_events_df['Latitude'].apply(make_floats)
all_events_df['Longitude'] = all_events_df['Longitude'].apply(make_floats)
```

```python
all_events_df['Latitude'].value_counts()
```

```python
#London_zones
# South East limits
se_limit_lat = 51.523811 # everthing lower than this value is south east
se_limit_lon = -0.101342 # everything greater than this value is south east

# East limits
e_limit_lat = 51.523811 # everything greater than this value is east
e_limit_lon = -0.022306 # everything greater than this is east

# South West limits
sw_limit_lat = 51.501631 # everthing lower than this value is south east
sw_limit_lon = -0.101342 # everything lower than this value is south east

# West limits
w_limit_lat = 51.501072  # everthing greater than this value is south east
w_limit_lon = -0.182406  # everthing lower than this value is south east

# Central limits
c_toplimit_lat = 51.501072 # everthing greater than this value is central
c_lowlimit_lat = 51.523811 # everything lower than this value is central

c_toplimit_lon = -0.101342 # everything lower than this value is central
c_lowlimit_lon = -0.182406 # everthing greater than this value is central

# North limits
n_toplimit_lon = -0.022306 # everything lower than this value is central
n_lowlimit_lon = -0.182406 # everthing greater than this value is central

n_lowlimit_lat = 51.523811 # everthing greater than this value is central
```

```python
london_loc = []
for i in range(len(all_events_df.index)):
    if all_events_df['Latitude'][i] == "London-wide":
        london_loc.append('London_all') # London wide
    elif all_events_df['Latitude'][i] > c_toplimit_lat and all_events_df['Latitude'][i]< c_lowlimit_lat and all_events_df['Longitude'][i]< c_toplimit_lon and all_events_df['Longitude'][i]>c_lowlimit_lon:
            london_loc.append('Central') # Central
    elif all_events_df['Longitude'][i] < n_toplimit_lon and all_events_df['Longitude'][i] > n_lowlimit_lon and all_events_df['Latitude'][i] > n_lowlimit_lat:
        london_loc.append('North') # North
    elif all_events_df['Longitude'][i] < w_limit_lon and all_events_df['Latitude'][i] > w_limit_lat:
        london_loc.append('West') # West
    elif all_events_df['Longitude'][i] < sw_limit_lon and all_events_df['Latitude'][i] < sw_limit_lat:
        london_loc.append('South_West')  # South_West
    elif all_events_df['Longitude'][i] > e_limit_lon and all_events_df['Latitude'][i] > e_limit_lat:
        london_loc.append('East') # East
    elif all_events_df['Longitude'][i] > se_limit_lon and all_events_df['Latitude'][i] < se_limit_lat:
        london_loc.append('South_East') # South East
    else:
        london_loc.append('Other')
```

```python
print(set(london_loc))
```

```python
all_events_df["London_zone"] =london_loc
```

```python
all_events_df.head()
```

```python
all_events_df["London_zone"].shape
```

```python
from sklearn.preprocessing import OneHotEncoder

print(f"The unique values for 'Zone' are {all_events_df.London_zone.unique()}")

# Instantiate the OneHotEncoder
ohe = OneHotEncoder(sparse = False) 

# Fit encoder
ohe.fit(all_events_df[['London_zone']]) 

# Display the detected categories
print(f"The categories detected by the OneHotEncoder are {ohe.categories_}")
```

```python
# Display the generated names
print(f"The column names for the encoded values are {ohe.get_feature_names_out()}")

# Transform the current "Street" column
all_events_df[ohe.get_feature_names_out()] = ohe.transform(all_events_df[['London_zone']])

# Drop the column "Street" which has been encoded
all_events_df.drop(columns = ["London_zone"], inplace = True)

# Show the dataset
all_events_df.head(3)
```
