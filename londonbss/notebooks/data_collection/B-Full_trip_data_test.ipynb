```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import io

def read_url(url):
    s=requests.get(url).content
    c=pd.read_csv(io.StringIO(s.decode('utf-8')))
    return c
```

```python
data=pd.read_csv('/Users/thammadoltansrivorarat/code/elsebasmar/london-bss/raw_data/tripcsvbame.csv')
```

```python
link="https://cycling.data.tfl.gov.uk/usage-stats/"

# e bike introduce at file 283
old_outlier=[]
e_bike=[]

for item in data['usage-stats/'][283:]:
    if item[-3:]=='csv':
        file=link+item[1:]
        if file[-5]=='2' or file[-5]=='3':
            e_bike.append(file)
        else:
            old_outlier.append(file)
        #frame.append(read_url(file))
```

```python
len(e_bike)
```

```python
frame=[]
for item in e_bike:
    frame.append(read_url(item))
```

```python
result1=pd.concat(frame)
result1
```

```python
result1=result1.drop(columns=['Bike model'],axis=1)
result1=result1.drop(columns=['Total duration'],axis=1)
```

```python
date_format = '%d/%m/%Y %H:%M:%S'
result1['End Date']=pd.to_datetime(result1['End Date'])
result1['Start Date']=pd.to_datetime(result1['Start Date'])
result1
```

```python
new_result1=result1.set_index('Start Date').groupby([pd.Grouper(freq='h'), 'StartStation Name', 'EndStation Name']).count()
new_result1=new_result1.reset_index()
new_result1
```
